# py-destiny — visualisation R⇄Py parity

`plot.DiffusionMap` + `plot.DPT` on Guo qPCR fixture.

## 1. Run R + Py drivers

```python
import subprocess
from pathlib import Path
PORT = Path('..').resolve(); OUT = PORT/'tests'/'plot_compare'
subprocess.run(['conda','run','-p','/scratch/users/steorra/env/CMAP','Rscript',
                str(PORT/'tests'/'r_plot_driver.R'), str(OUT)], check=True, cwd=PORT)
subprocess.run(['python', str(PORT/'tests'/'py_plot_driver.py'), str(OUT)], check=True, cwd=PORT)
print('Renders ready')
```

## 2. plot.DPT on Guo data

```python
from PIL import Image
import matplotlib.pyplot as plt
fig, (a1, a2) = plt.subplots(1, 2, figsize=(13, 5))
a1.imshow(Image.open('../tests/plot_compare/R_dpt.png')); a1.set_title('R reference'); a1.axis('off')
a2.imshow(Image.open('../tests/plot_compare/Py_dpt.png')); a2.set_title('Py port'); a2.axis('off')
plt.suptitle('plot(DPT(dm))'); plt.tight_layout(); plt.show()
```

Identical Y-branch topology in DC1×DC2 space. Both ports show the same pseudotime gradient from the 1-cell timepoint (dark, low DPT) to mature lineages (light, high DPT).
